```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
%matplotlib inline
```

### å›³1.2

```python
x = np.linspace(0, 1, 100)
y = np.sin(2 * math.pi * x)
plt.plot(x, y, label='Graph')
plt.grid(True)
x_point = 0.0
for i in range(10):
    y_point = np.sin(2 * math.pi * x_point) + random.uniform(-0.2, 0.2)
    plt.scatter(x_point, y_point, c='red', label='Points', alpha=0.5)
    x_point += 0.1
plt.show()
```
